```python
#Projeto 2: Calculadora

def soma (v):
    l=0
    for i in v:
        l=l+i
    print(l)
    return l

def sub (v):
    lista=[]
    k=1
    for i in v:
        lista.append(i)
        
    r=lista[0]
    
    while k < len(lista):    
        r=r-lista[k]
        k=k+1
    else:
        print(r)
        return r

def mult (v):
    l=1
    for i in v:
        l*=i
    print(l)
    return l

def div (v):
    lista=[]
    k=1
    for i in v:
        lista.append(i)
        
    r=lista[0]
    
    while k < len(lista):    
        r=r/lista[k]
        k=k+1
    else:
        print(r)
        return r
    


adiciona=(lambda *var: soma(var))
subtrai=(lambda *var: sub(var))
multiplica=(lambda *var: mult(var))
dividi=(lambda *var: div(var))

#Queria otimizar essa parte pois as fótmulas conseguem receber várias variáveis mas de uma forma visual não consegui
#resolver, fiquei meio sem ideias


stop=input('Digite stop para parar de calcular e sim para continuar' )

while stop != 'stop':
    
    a, b, c = input('informe a conta na calculadora: ').split(" ")

    if b=='mais':
        adiciona(float(a),float(c))
    elif b =='menos':
        subtrai(float (a),float (c))
    elif b =='vezes':
        multiplica(float (a),float (c))
    elif b =='dividido':
        dividi(float (a),float (c))
    else:
        print('Conta escrita errada')
        
    stop=input('Digite stop para parar de calcular e sim para continuar' )
```
